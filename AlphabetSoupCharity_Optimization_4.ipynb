```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dropout, Dense, BatchNormalization, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l1
from tensorflow.keras.callbacks import ModelCheckpoint

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME', AND 'SPECIAL_CONSIDERATIONS'.
#  List of columns to drop
columns_to_drop = ['EIN', 'NAME', 'SPECIAL_CONSIDERATIONS']

# Drop the columns from application_df
application_df = application_df.drop(columns_to_drop, axis=1)
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
value_counts = application_df['APPLICATION_TYPE'].value_counts()
cutoff_value = 500
application_types_to_replace = value_counts[value_counts < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts().loc[lambda x : x > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
value_counts_classification = application_df['CLASSIFICATION'].value_counts()
cutoff_value_classification = 1000
classifications_to_replace = value_counts_classification[value_counts_classification < cutoff_value_classification].index.tolist()
```

```python
# Replace in dataframe
application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(classifications_to_replace, "Other")
# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, dtype=float)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns=['IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

# An input layer with 41 neurons.
# A hidden layer with 128 neurons and ReLU activation.
# A hidden layer with 64 neurons and LeakyReLU activation.
# A hidden layer with 32 neurons and ReLU activation.
# An output layer with 1 neuron and sigmoid activation.

nn = tf.keras.Sequential([
    Dense(128, activation='relu', input_dim=41, kernel_regularizer=l1(0.0001)),  # Add L1 regularization (adjust value)
    BatchNormalization(),
    Dropout(0.2),
    Dense(64, activation=LeakyReLU()),
    BatchNormalization(),
    Dense(32, activation='relu', kernel_regularizer=l1(0.0001)),  # Add L1 regularization (adjust value)
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
# Check the structure of the model
nn.summary()
```

```python
# Compile the model

nn.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), metrics=['accuracy'])

# Early stopping
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

# Include the epoch in the file name (uses f-strings)
checkpoint_path = f"weights_epoch{{epoch:02d}}.hdf5"

# Create a callback that saves the model's weights every 5 epochs
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    period=5,
)

# Train the model with the new callback
nn.fit(
    X_train_scaled,
    y_train,
    epochs=100,
    batch_size=64,
    validation_data=(X_test_scaled, y_test),
    callbacks=[early_stop, cp_callback],
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
